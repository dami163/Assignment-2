```python
import cv2 as cv
import numpy as np
import matplotlib.pylab as plt

img = cv.imread('Crop_field_cropped.jpg', cv.IMREAD_GRAYSCALE)

edges = cv.Canny(img,550,690)
indices = np.where(edges != [0])
x = indices[1]
y = indices[0]

plt.scatter(x,y , s=10 , c='r')
plt.title('scatter plot')
plt.show()
```
